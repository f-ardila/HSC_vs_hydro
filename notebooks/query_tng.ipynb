```python
from __future__ import division
import requests

baseUrl = 'http://www.tng-project.org/api/'
headers = {"api-key":"e05ffb2ca14084c00423a3cd77f7e268"}

def get(path, params=None):
    # make HTTP GET request to path
    r = requests.get(path, params=params, headers=headers)

    # raise exception if response code is not HTTP SUCCESS (200)
    r.raise_for_status()

    if r.headers['content-type'] == 'application/json':
        return r.json() # parse json responses automatically
    return r
```

```python
def print_completeness_values(i, snap_n):
    
    r = get(baseUrl)
    sim = get( r['simulations'][i]['url'] )
    print(sim['name'])
    snaps = get( sim['snapshots'] )
    snap = get( snaps[snap_n]['url'] )
#     groupcat = snap['files']['groupcat']
    
#     groupcat_query = "?Subhalo=Group_M_Crit200"
    
    mass_min = 10**13 / 1e10 * 0.704
    mass_max = 10**(13.42) / 1e10 * 0.704
    mass_stars_min = 10**(11.2) / 1e10 * 0.704
    
    #our sample
    search_query1 = "?mass_stars__gt="+ str(mass_stars_min)+ "&primary_flag=1"
    
    #entire sample
    search_query2 = "?mass__gt=" + str(mass_min)+ "&primary_flag=1" 
    search_query3 = "?mass_stars__gt="+ str(mass_stars_min)+ "&primary_flag=1"+ "&mass__gt=" + str(mass_min)
    
    #first halo mass bin
    search_query4 = "?mass__gt=" + str(mass_min)+ "&primary_flag=1" +"&mass__lt=" + str(mass_max)
    search_query5 = "?mass_stars__gt="+ str(mass_stars_min)+ "&primary_flag=1"+ "&mass__gt=" + str(mass_min)+"&mass__lt=" + str(mass_max)
    
    
    n_halos_list = []
    for search_query in [search_query1, search_query2, search_query3, search_query4, search_query5]:
        url = snap['subhalos'] + search_query
        subhalos = get(url)
        n_halos = subhalos['count']
        n_halos_list.append(n_halos)
        
        print(n_halos)

    print('{0} completeness over entire sample'.format(round(n_halos_list[2]/n_halos_list[1],3)))
    print('{0} completeness over first halo mass bin'.format(round(n_halos_list[4]/n_halos_list[3],3)))
    
    return subhalos
    
```
