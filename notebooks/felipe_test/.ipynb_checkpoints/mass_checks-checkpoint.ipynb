# Compare masses:
    * catalog mass
    * summed pixels mass
    * mass from fit

```python
%matplotlib inline 
%load_ext autoreload 
%autoreload 2

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)

import pickle
```

```python
#open pickels
illustris_pickle='/Users/fardila/Documents/GitHub/HSC_vs_hydro/Data/Illustris/Illustris.pkl'
tng_pickle='/Users/fardila/Documents/GitHub/HSC_vs_hydro/Data/TNG/TNG.pkl'

pkl = open(illustris_pickle,'rb')  
illustris = pickle.load(pkl)  
pkl.close()

pkl = open(tng_pickle,'rb')  
tng = pickle.load(pkl)  
pkl.close()
```

```python
def get_masses(array):
    cat_masses=np.array([i[1] for i in array])
    summed_masses=np.array([i[2] for i in array])
    fit_masses=np.array([i[0]['mag_tot'][0]/-2.5 for i in array])
    
    return [cat_masses, summed_masses, fit_masses]
```

# Illustris

```python
masses_illustris = get_masses(illustris)
```

```python
plt.hist(masses_illustris[0], bins=15, histtype='step', linewidth=4, label='Catalog mass')
plt.hist(masses_illustris[1], bins=15, histtype='step', linewidth=4, label='Summed mass')
plt.hist(masses_illustris[2], bins=15, histtype='step', linewidth=4, label='Fit mass')

plt.legend()
plt.xlabel('log Mass')
plt.show()
```

```python
#mass vs. mass
plt.scatter(masses_illustris[0],masses_illustris[2], marker='.', s=10)
plt.plot([11,12.5],[11,12.5])
plt.xlabel('Catalog Mass')
plt.ylabel('Fit Mass')
plt.show()
```

```python
# differences
illustris_diff_cat_sum=masses_illustris[0]-masses_illustris[1]
illustris_diff_cat_fit=masses_illustris[0]-masses_illustris[2]

plt.hist(illustris_diff_cat_sum, bins=15, histtype='step', linewidth=4,
         label='Catalog - Sum : mean ' +str(round(np.mean(illustris_diff_cat_sum),3))+\
         '; std '+str(round(np.std(illustris_diff_cat_sum),3)))
plt.hist(illustris_diff_cat_fit, bins=15, histtype='step', linewidth=4,
         label='Catalog - Fit : mean ' +str(round(np.mean(illustris_diff_cat_fit),3))+\
         '; std '+str(round(np.std(illustris_diff_cat_fit),3)))

plt.legend()
plt.xlabel('Mass Difference')
plt.show()
```

```python
plt.scatter(masses_illustris[0],illustris_diff_cat_sum, s=10, label='Catalog - Sum')
plt.scatter(masses_illustris[0],illustris_diff_cat_fit, s=10, label='Catalog - Fit')

plt.legend()
plt.xlabel('Catalog Mass')
plt.ylabel('Mass Difference')
plt.show()
```

```python
bins_cat= np.linspace(11.2, 12.5, 10)
db=(bins_cat[1]-bins_cat[0])/2.
bin_centers=bins_cat[:-1]+ db

#assign each mass to a bin
digitized = np.digitize(masses_illustris[0], bins_cat)

#means
bin_means_illustris_diff_cat_sum = np.array([illustris_diff_cat_sum[digitized == i].mean() for i in range(1,len(bins_cat))])
bin_means_illustris_diff_cat_fit = np.array([illustris_diff_cat_fit[digitized == i].mean() for i in range(1, len(bins_cat))])

#stds
bin_std_illustris_diff_cat_sum = np.array([illustris_diff_cat_sum[digitized == i].std() for i in range(1, len(bins_cat))])
bin_std_illustris_diff_cat_fit = np.array([illustris_diff_cat_fit[digitized == i].std() for i in range(1, len(bins_cat))])

# #change format to make plots of width
# new_bins=[]
# new_means_diff_cat_sum=[]
# for i, bini in enumerate(bin_centers):
#     new_bins.append(bini-db)
#     new_bins.append(bini)
#     new_bins.append(bini+db)
    
#     #append 3 times
#     new_means_diff_cat_sum.append(bin_means_illustris_diff_cat_sum[i])
#     new_means_diff_cat_sum.append(bin_means_illustris_diff_cat_sum[i])
#     new_means_diff_cat_sum.append(bin_means_illustris_diff_cat_sum[i])
```

```python
#plot
plt.plot(bin_centers,bin_means_illustris_diff_cat_sum, label='Catalog - Sum')
plt.plot(bin_centers,bin_means_illustris_diff_cat_fit, label='Catalog - Fit')

plt.fill_between(bin_centers,bin_means_illustris_diff_cat_sum-bin_std_illustris_diff_cat_sum,
                bin_means_illustris_diff_cat_sum+bin_std_illustris_diff_cat_sum, alpha=0.5)
plt.fill_between(bin_centers,bin_means_illustris_diff_cat_fit-bin_std_illustris_diff_cat_fit,
                bin_means_illustris_diff_cat_fit+bin_std_illustris_diff_cat_fit, alpha=0.5)

# plt.scatter(masses_illustris[0],illustris_diff_cat_sum, s=10, label='Catalog - Sum')
# plt.scatter(masses_illustris[0],illustris_diff_cat_fit, s=10, label='Catalog - Fit')

plt.legend()
plt.xlabel('Catalog Mass')
plt.ylabel('Mass Difference')
plt.show()
```
